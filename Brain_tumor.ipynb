```python
results = model.train(
    data='/content/drive/MyDrive/Project/brain-tumor/brain-tumor.yaml',
    epochs=80,          # More epochs = better learning
    batch=32,           # Avoids memory crash, stable training
    imgsz=640,

    # --- Augmentations ---
    scale=0.8,          # Better for medical images
    hsv_h=0.0,          # Disable color distortions (MRI are grayscale)
    hsv_s=0.0,
    hsv_v=0.0,
    mosaic=0.5,         # Mild mosaic helps small datasets
    mixup=0.0,          # Avoid mixup for medical images
    copy_paste=0.0,     # Medical imgs don’t benefit from this

    # --- Training Optimizer ---
    lr0=0.001,          # Lower learning rate = smoother training
    patience=40,        # Stop early if no improvement

    device="0",
)
```

```python
from ultralytics import YOLO

# Load model
model = YOLO("Sample/best.pt")

# Path to test image
img_path = "Sample/test_images/64 (5).jpg"

# Run inference
results = model(img_path)

# Show results
results[0].show()

# Save predicted image
results[0].save(filename="output.jpg")


```
